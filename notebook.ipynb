```python
# transform() function, fills null values for numerical columns and drops columns which are not required
def transform(raw_data):
  # Write your code here
    raw_data['Weekly_Sales'].fillna(raw_data['Weekly_Sales'].mean(), inplace = True)
    raw_data['CPI'].fillna(raw_data['CPI'].mean(), inplace = True)
    raw_data['Unemployment'].fillna(raw_data['Unemployment'].mean(), inplace = True)
    raw_data['Size'].fillna(raw_data['Size'].mean(), inplace = True)
    raw_data['Month'] = raw_data['Date'].dt.month
    
    clean_data = raw_data.loc[merged_df['Weekly_Sales']>10000]
    required_columns =  [ "Store_ID",
        "Month",
        "Dept",
        "IsHoliday",
        "Weekly_Sales",
        "CPI",
        "Unemployment"
         ]
    clean_data = clean_data[required_columns]
    return clean_data
```

```python
# Calling the transform() function 
clean_data = transform(merged_df)
```

```python
# A function that returns a aggregated values dataframe
def avg_weekly_sales_per_month(clean_data):
    # Write your code here
    agg_data = clean_data.groupby('Month')['Weekly_Sales'].mean().reset_index().round(2)
    agg_data.rename(columns = {'Weekly_Sales':'Avg_Sales'}, inplace = True)
    return agg_data
```

```python
# Aggregate data
agg_data = avg_weekly_sales_per_month(clean_data)
```

```python
#  load() function that takes in the cleaned DataFrame and the aggregated one with the paths where they are going to be stored
def load(full_data, full_data_file_path, agg_data, agg_data_file_path):
    full_data.to_csv(full_data_file_path, index = False)
    agg_data.to_csv(agg_data_file_path, index = False)
    # Write your code here
```
